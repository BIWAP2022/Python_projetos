```python
import pandas as pd
```

```python
dados = pd.read_csv("New Report.csv",parse_dates=['Base_Data'])
```

```python
dados2 = dados.groupby(["Base_Data", 'Base_Codigo'], as_index=False)['Base_Qtd'].sum()
dados2.head()
```

```python
dados2['Dia_do_mes'] = dados2['Base_Data'].dt.day
dados2['Mes'] = dados2['Base_Data'].dt.month
dados2['Ano'] = dados2['Base_Data'].dt.year
dados2['Dia_da_semana'] = dados2['Base_Data'].dt.weekday
dados2.head()
```

```python
dados3 = dados2.drop("Base_Data", axis=1)


from sklearn.tree import DecisionTreeRegressor

modelo = DecisionTreeRegressor()
modelo.fit(dados3[['Dia_do_mes','Mes', 'Ano', 'Dia_da_semana', 'Base_Codigo']], dados3['Base_Qtd'])


futuro = pd.date_range("2022-05-01", "2022-08-31", freq='D')

futuro_todos = []

for Codigo in dados3['Base_Codigo'].unique():
    futuro_df = pd.DataFrame()
    futuro_df['Dia_do_mes'] = futuro.day
    futuro_df['Mes'] = futuro.month
    futuro_df['Ano'] = futuro.year
    futuro_df['Dia_da_semana'] = futuro.weekday
    futuro_df['Base_Codigo'] = Codigo     
    p = modelo.predict(futuro_df)
    
    futuro_df['Base_Qtd'] = p
    
    futuro_todos.append(futuro_df)
    
futuro_todos_df = pd.concat(futuro_todos, ignore_index=True)
futuro_todos_df.head()
```

```python
futuro_todos_df.to_csv('futuro_todos_df.csv')
```
